```python
import pandas as pd
import numpy as np
import json
import pprint
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db = client.companies
```

```python
data = pd.read_json('68_Berlin_Munich.json', orient = 'records', encoding = 'utf8')
data.head()
data.columns
```

```python
comp = data[['name', 'offices', 'funding_rounds', 'total_money_raised', 'founded_year', 'number_of_employees', 'category_code', 'description', 'overview', 'tag_list']].copy()
comp
```

```python
def get_coords(data):
    
    res = {
        "city": None
        "lat": None
        "lng": None
        "site": None
    }

    data = data['offices']
    oficinas = data
    
    for oficina in oficinas: 
        if oficina.get('country_code') == 'DEU' and oficina.get('latitude') and oficina.get('longitude'): 
            
            coords = {
            "type": "Point",
            "coordinates": [data[0]['longitude'], data[0]['latitude']]
            }

    return res = {
        "city": oficina.get('city')
        "lat": oficina.get('latitude')
        "lng": oficina.get('longitude')
        "site": coords
    }
        break

location = comp[['offices']].apply(get_coords, result_type='expand', axis=1)
```

```python
expanded = pd.concat([comp,location], axis=1)[["name","city","total_money_raised","site","lat","lng","founded_year","number_of_employees","category_code","description","overview","tag_list"]]
expanded.head(50)
```

```python
#https://docs.mongodb.com/manual/reference/operator/query/near/index.html
def findNear(geopoint, radio_max_meters=1000):
    return db.first_office.find({
        "oficina_principal": {
         "$near": {
           "$geometry": geopoint,
           "$maxDistance": radio_max_meters,
         }
       }
    })


# https://developers.google.com/maps/documentation/geocoding/intro
park_avenue_con_21_street = {
    "type":"Point",
    "coordinates":[-73.987308,40.738935]
}
radio_max_meters = 10000
num_offices = findNear(park_avenue_con_21_street, radio_max_meters).count()

print(f"Hay {num_offices} oficinas cerca a {radio_max_meters} metros")
print("La mas cercana es")
print(list(findNear(park_avenue_con_21_street, radio_max_meters).limit(1)))
```
