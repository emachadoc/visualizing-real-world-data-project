```python
import pandas as pd
import numpy as np
import json
import pprint
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db = client.companies
```

```python
# El objetivo de mi empresa es abrir una sucursal en Alemania, por lo que filtro la base de datos
# para quedarme exclusivamente con los registros con oficinas en ese país

german_comps = db.companies.find({
    "offices.country_code": 'DEU'
        })
```

```python
# Creo el DataFrame

euro_comps = pd.DataFrame(german_comps)
euro_comps.head()
```

```python
# Selecciono las columnas que me interesan y descarto las demás

less_columns = euro_comps[['name', 'offices', 'funding_rounds', 'total_money_raised', 'founded_year', 'number_of_employees', 'category_code', 'description', 'overview', 'tag_list']].copy()
less_columns.head()
```

```python
# Comprobación de tamaño del DataFrame actual

less_columns.shape
```

```python
# Comprobación de los subsets dentro de offices

print(less_columns['offices'][0])
```

```python
# Extraigo los datos de ciudad, latitud, longitud y geo para cada oficina de las ubicadas en Alemania y
# que tengan latitud y longitud

def get_coords(less_columns):
    # gets a row from less_columns df and returns the first german office in dict shape
    res = {
        "city": None,
        "lat": None,
        "lng": None,
        "geo": None
    }

    oficinas = less_columns['offices']
    
    for oficina in oficinas: 
        
        if oficina.get('country_code') == 'DEU' and oficina.get('latitude') and oficina.get('longitude'): 
            
            coords = {
            "type": "Point",
            "coordinates": [oficina.get('longitude'), oficina.get('latitude')]
            }
            res = {
                "city": oficina.get('city'),
                "lat": oficina.get('latitude'),
                "lng": oficina.get('longitude'),
                "geo": coords
                }
            break
    return res

location = less_columns[['offices']].apply(get_coords, result_type='expand', axis=1)
location.head()
```

```python
# Creo un nuevo DataFrame 'germany' agregando las columnas anteriores 'location' al anterior DataFrame 'less-columns'

germany = pd.concat([less_columns,location], axis=1)[["name","city","offices","geo","lat","lng","total_money_raised","founded_year","number_of_employees","category_code","description","overview","tag_list"]]
germany.head(3)
```

```python
# Elimino los registros en los que no tengo datos de latitud (ni longitud), ya que no podré representarlos

germany.dropna(subset=['lat'], inplace=True)
germany.head(3)
```

```python
# Ordeno descendentemente por año de fundación

germany.sort_values(by=['founded_year'], ascending=False, inplace=True)
germany.head(3)
```

```python
# Exporto a .csv

germany.to_csv('germany.csv')
```

```python
# Exporto a .json

germany.to_json('germany.json', orient='records')
```
